APPLY WHERE FILTER (REGION ,CATEGORY , DATES AND SALES )

```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect("sales.db")
```

```python
query="""
SELECT *
FROM sales
WHERE Region='East';
"""
pd.read_sql(query, conn)
```

```python
query="""
SELECT *
FROM sales
WHERE Sales>500;
"""
pd.read_sql(query,conn)
```

```python
query="""
SELECT *
FROM sales 
WHERE Category = 'Technology';
"""
pd.read_sql(query,conn)
```

USE GROUP BY FOR AGGREGATIONS 

```python
query="""
SELECT Category,
SUM(Sales)
FROM sales 
GROUP BY Category
"""
pd.read_sql(query,conn)
```

```python
query="""
SELECT Region, AVG(Sales)
FROM sales
GROUP BY Region 
"""
pd.read_sql(query,conn)
```

```python
query=""" 
SELECT Category, SUM(Quantity)
FROM sales
GROUP BY Category
"""
pd.read_sql(query,conn)
```

```python
query=""" 
SELECT *
FROM sales
ORDER BY Sales DESC
LIMIT 10"""
pd.read_sql(query,conn)
```

```python
query = """
SELECT [Product Name],
       SUM(Sales) AS total_sales
FROM sales
GROUP BY [Product Name]
ORDER BY total_sales DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT Category,
       SUM(Sales) AS total_sales
FROM sales
GROUP BY Category
ORDER BY total_sales DESC
"""

pd.read_sql(query, conn)
```

Solve use Cases 

```python
query = """
SELECT strftime('%Y-%m', [Order Date]) AS month,
       SUM(Sales) AS total_sales
FROM sales
GROUP BY month
ORDER BY month
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT [Customer Name],
       SUM(Sales) AS total_sales
FROM sales
GROUP BY [Customer Name]
ORDER BY total_sales DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT [Row ID],
       COUNT(*) AS duplicate_count
FROM sales
GROUP BY [Row ID]
HAVING COUNT(*) > 1
"""

pd.read_sql(query, conn)
```

VALIDATE RESULTS(Rows Count ,data quality  )

```python
query="""
SELECT COUNT(*) AS total_rows
FROM sales
"""
pd.read_sql(query,conn)
```

DATA QUALITY 

```python
query="""
SELECT *
FROM sales
WHERE Sales<0
"""
pd.read_sql(query,conn)
```

```python
query="""
SELECT *
FROM sales
WHERE Sales IS NULL
"""
pd.read_sql(query,conn)
```
