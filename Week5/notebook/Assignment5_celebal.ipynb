Objective:
To understand Spark fundamentals and perform data cleaning, transformation, filtering and aggregation using Spark DataFrames.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
```

```python
spark = SparkSession.builder \
    .appName("Spark Assignment") \
    .getOrCreate()

print("Spark Session Created Successfully")
```

```python
df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "false") \
    .option("multiLine", "true") \
    .option("quote", '"') \
    .option("escape", '"') \
    .csv(r"C:\Users\Acer\Downloads\Sample - Superstore.csv")
```

Dataset Exploration

```python
df.show(5,truncate = False)
```

```python
print("Columns:")
print(df.columns)
```

```python
from pyspark.sql.functions import col

df = df.withColumnRenamed("Sales", "TotalSales")

df = df.withColumn("TotalSales", col("TotalSales").cast("double"))
df = df.withColumn("Quantity", col("Quantity").cast("int"))
df = df.withColumn("Discount", col("Discount").cast("double"))
df = df.withColumn("Profit",col("Profit").cast("double"))
```

```python
df.printSchema()
```

```python
print("Total Rows:", df.count())
```

Data Cleaning

```python
df = df.dropDuplicates()

print("Rows after removing duplicates:")
print(df.count())
```

```python
df = df.dropDuplicates()
```

```python
df = df.na.drop()
```

```python
print("Rows After Cleaning:", df.count())
```

## Spark DataFrame Immutability

Spark DataFrames are immutable. Any transformation such as filter(),
withColumn(), select(), or dropDuplicates() creates a new DataFrame
instead of modifying the existing DataFrame.

Transformations

```python
df = df.withColumnRenamed("Sales", "TotalSales")
```

```python
df = df.withColumn(
    "TotalSales",
    col("TotalSales").cast("double")
)

df = df.withColumn(
    "Quantity",
    col("Quantity").cast("int")
)

df = df.withColumn(
    "Discount",
    col("Discount").cast("double")
)
```

```python
df.printSchema()
```

Filter Data

```python
df.filter(df.Category == "Furniture").show()
```

```python
df.filter(df.Region == "West").show()
```

```python
df.filter(df.TotalSales > 500).show()
```

```python
df.filter(df.Quantity >= 3).show()
```

Aggregation

```python
df.select(avg("TotalSales")).show()
```

```python
df.select(sum("TotalSales")).show()
```

```python
df.select(min("TotalSales")).show()
```

```python
df.select(max("TotalSales")).show()
```

GroupBy

```python
df.groupBy("Category").sum("TotalSales").show()
```

```python
df.groupBy("Region").avg("Profit").show()
```

```python
df.groupBy("Sub-Category").count().show()
```

```python
result = (
    df
    .filter(df.TotalSales > 500)
    .groupBy("Category")
    .sum("TotalSales")
)
```

```python
result.show()
```

```python
result.toPandas().to_csv(
    "results.csv",
    index=False
)
```
