Step 2: Perform Required Queries 

Write and execute SQL queries for each of the following: 

1.Find all orders where sales are greater than the average sales. (Subquery)  

2.Find the highest sales order for each customer. (Subquery)  

3.Calculate total sales for each customer. (CTE)  

4.Find customers whose total sales are above average. (CTE + Subquery)  

5.Rank all customers based on total sales. (Window Function)  

6.Assign row numbers to each order within a customer. (Window Function + PARTITION BY)  

7.Display top 3 customers based on total sales. (Window Function)  

 

1.) Find all orders where sales are greater than the average sales. (Subquery) 

```python
pd.read_sql(
    """SELECT * 
    FROM orders
    WHERE sales>(
      SELECT AVG(Sales)
      FROM orders );
      """,conn)
```

2. Find the highest sales order for each customer (Subquery)

```python
pd.read_sql("""
            SELECT customer_id, order_id,Sales
            FROM orders o
            WHERE sales=
            (
               SELECT MAX(SALES)
               FROM orders
               WHERE customer_id=o.customer_id
            );
            """,conn)
```

3. Calculate total sales for each customer (CTE)

```python
pd.read_sql("""
WITH customer_sales AS
(SELECT
        customer_id,
        SUM(Sales) AS total_sales
 FROM orders
 GROUP BY customer_id
)

SELECT *
FROM customer_sales
ORDER BY total_sales DESC; """,conn)
```

4. Find customers whose total sales are above average (CTE + Subquery)

```python
pd.read_sql("""
WITH customer_sales AS 
(
            SELECT customer_id, 
            SUM(Sales) AS total_sales
            FROM orders
            GROUP BY customer_id
)
SELECT *
FROM customer_sales
WHERE total_sales>
            (
            SELECT AVG(total_sales)
            FROM customer_sales
            );
            """,conn)
```

5. Rank all customers based on total sales (Window Function)

```python
pd.read_sql("""
WITH customer_sales AS 
(
    SELECT customer_id , SUM(Sales) AS total_sales
    FROM orders 
    GROUP BY customer_id
) 
SELECT customer_id , total_sales, 
RANK() OVER( ORDER BY total_sales DESC) AS customer_rnk
FROM customer_sales;""",conn)
```

6. Assign row numbers to each order within a customer (PARTITION BY)

```python
pd.read_sql("""
       SELECT customer_id , order_id, Sales,
       ROW_NUMBER() OVER(
              PARTITION BY customer_id
              ORDER BY Sales DESC) AS row_num
       FROM orders;
       """,conn)
```

7. Display Top 3 Customers Based on Total Sales (Window Function)

```python
pd.read_sql("""
WITH customer_sales AS 
(
  SELECT customer_id, SUM(Sales) AS total_sales
  FROM orders
  GROUP BY customer_id
),
ranked_customers AS
(
    SELECT customer_id, total_sales,
    RANK() OVER(
            ORDER BY total_sales DESC) AS sales_rank 
    FROM customer_sales
)
SELECT *
FROM  ranked_customers
WHERE sales_rank<=3;

""",conn)
```

Step 3: Final Combined Query 

Write one final query that shows:

* FINAL COMBINED QUERY

```python
final_result= pd.read_sql("""
WITH customer_sales AS
(
    SELECT c.customer_id, c.customer_name, SUM(o.Sales) AS total_sales
    FROM customers c
    JOIN orders o 
       ON c.customer_id=o.customer_id
    GROUP BY c.customer_id, c.customer_name
)
SELECT customer_name, total_sales,
      RANK() OVER(
                  ORDER BY total_sales DESC ) AS customer_rank 
FROM customer_sales;                          
""",conn)
final_result.head(20)
```

Customers with Rank 1 generate the highest total sales and are the most valuable customers for the business.

Mini Project: Customer Sales Insights 

Answer the following using SQL: 

1.)Who are the top 5 customers?  

2.)Who are the bottom 5 customers?  

3.)Which customers made only one order?  

4.)Which customers have above-average sales?  

5.)What is the highest order value per customer? 

1.)Who are the top 5 customers?
Insight:

The top 5 customers contribute the highest revenue and represent the most valuable customer segment.

```python
pd.read_sql("""
WITH customer_sales AS 
(
    SELECT c.customer_id, c.customer_name, SUM(o.Sales) AS total_sales
    FROM customers c
    JOIN orders o 
        ON c.customer_id = o.customer_id
    GROUP BY c.customer_id, c.customer_name
)
SELECT *
FROM customer_sales
ORDER BY total_sales DESC 
LIMIT 5;
""",conn)
```

2.)Who are the bottom 5 customers?


```python
pd.read_sql("""
WITH customer_sales AS
(
            SELECT c.customer_id, c.customer_name, SUM(o.Sales) AS total_sales 
            FROM customers c
            JOIN orders o 
                ON c.customer_id=o.customer_id
            GROUP BY c.customer_id, c.customer_name
)
SELECT *
FROM customer_sales 
ORDER BY total_sales ASC
LIMIT 5; 
""",conn)
```

3.)Which customers made only one order? 
Insight:

These customers may be targeted through retention campaigns to encourage repeat purchases.

```python
pd.read_sql("""SELECT c.customer_id, c.customer_name, COUNT(o.order_id) AS total_orders
               FROM customers c
               JOIN orders o 
                   ON c.customer_id = o.customer_id 
               GROUP BY c.customer_id, c.customer_name
               HAVING COUNT(o.order_id)=1;
            """,conn)
```

4.)Which customers have above-average sales? 
Insight:

Customers above the average sales threshold are strong candidates for loyalty and premium programs.

```python
pd.read_sql("""
WITH customer_sales AS 
(
    SELECT c.customer_id , c.customer_name,SUM(o.Sales) AS total_sales 
    FROM customers c 
    JOIN orders o
        ON c.customer_id = o.customer_id
    GROUP BY c.customer_id, c.customer_name
)
SELECT *
FROM customer_sales 
WHERE total_sales>
    (
            SELECT AVG(total_sales)
            FROM customer_sales
    )
ORDER BY total_sales DESC;
""",conn)
```

5.)What is the highest order value per customer?

```python
pd.read_sql("""
SELECT c.customer_id, c.customer_name, MAX(o.Sales) AS highest_order_value
FROM customers c
JOIN orders o
    ON c.customer_id= o.customer_id
GROUP BY c.customer_id, c.customer_name
ORDER BY highest_order_value DESC;
""",conn)
```
